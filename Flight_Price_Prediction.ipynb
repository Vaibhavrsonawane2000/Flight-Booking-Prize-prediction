```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/content/Flight_Booking.csv')
df
```

```python
df.info()
df.describe()
df.shape
df.isnull().sum()
```

```python
#Airlinw VS Price
plt.figure(figsize=(15,5))
sns.lineplot(x=df['airline'],y=df['price'],color='red')
plt.title('Airline VS Price')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.show()
```

```python
#Day left departure VS Ticket price
plt.figure(figsize=(15,5))
sns.lineplot(x=df['days_left'],y=df['price'],color='red')
plt.title('Day left departure VS Ticket price')
plt.xlabel('Day left departure')
plt.ylabel('Price')
plt.show()
```

```python
#Price range of all flight
plt.figure(figsize=(15,5))
sns.barplot(x=df['airline'],y=df['price'])
plt.title('Price range of all flight')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.show()
```

```python
#Range of price of all flight of economy and Business class
plt.figure(figsize=(15,5))
sns.barplot(x=df['class'],y=df['price'],hue=df['airline'])
plt.title('Range of price of all flight of economy and Business class')

```

```python
#Range of Price of flight of source and destination coty according to the day left
fig,ax=plt.subplots(1,2,figsize=(15,5))
sns.lineplot(x=df['days_left'],y=df['price'],hue=df['source_city'],ax=ax[0])
sns.lineplot(x=df['days_left'],y=df['price'],hue=df['destination_city'],ax=ax[1])
plt.show()
```

```python
#Visulization of categorical features with countplot
plt.figure(figsize=(15,5))

plt.subplot(4,2,1)
sns.countplot(x=df['airline'],data=df)
plt.title('frequency of Airline')

plt.subplot(4,2,2)
sns.countplot(x=df['source_city'],data=df)
plt.title('frequency fo source_city')

plt.subplot(4,2,3)
sns.countplot(x=df['destination_city'],data=df)
plt.title('frequency of destination_city')

plt.subplot(4,2,4)
sns.countplot(x=df['departure_time'],data=df)
plt.title('frequency of departure_time')

plt.subplot(4,2,5)
sns.countplot(x=df['arrival_time'],data=df)
plt.title('frequency of Arrival time')

plt.subplot(4,2,6)
sns.countplot(x=df['stops'],data=df)
plt.title('frequency of stops')

plt.subplot(4,2,7)
sns.countplot(x=df['class'],data=df)
plt.title('frequency of class')
plt.show()
```
